```python

df['PassengerId'] = pd.to_numeric(df['PassengerId'], errors='coerce', downcast='integer')
df['Pclass'] = pd.to_numeric(df['Pclass'], errors='coerce', downcast='integer')

df['Survived'] = pd.to_numeric(df['Survived'], errors='coerce', downcast='integer')

df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')

df['SibSp'] = pd.to_numeric(df['SibSp'], errors='coerce', downcast='integer')
df['Parch'] = pd.to_numeric(df['Parch'], errors='coerce', downcast='integer')

df['Name'] = df['Name'].astype(str)
df['Sex'] = df['Sex'].astype('category')
df['Ticket'] = df['Ticket'].astype(str)
df['Cabin'] = df['Cabin'].astype(str)
df['Embarked'] = df['Embarked'].astype('category')

print(df.dtypes)
```

VISUALIZATION


```python
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
df = pd.read_csv("tested.csv")
```

 Histogram of Age

```python
plt.figure(figsize=(8, 6))
sn.histplot(df["Age"] , bins=30 , kde=True)
plt.title("Age Distribution")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Box Plot of Fare

```python
plt.figure(figsize=(8,6))
sn.histplot(df["Fare"] , bins=30 ,kde=True)
plt.title("Fare DISTRIBUTION")
plt.xlabel("FARE")
plt.ylabel("FREQUENCY")
plt.show()
```

Bar Plot of Survival by Passenger Class

```python
plt.figure(figsize=(8,6))
sn.barplot(x="Pclass" , y="Survived" , data=df , ci=None)
plt.title("Survival by Class")
plt.xlabel("PASSENGER CLASS")
plt.ylabel("SURVIVED")
plt.show()
```
